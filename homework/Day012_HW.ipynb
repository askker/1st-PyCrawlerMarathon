### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
#print(soup)
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
news={}
for d in soup.find(class_="part_list_2").find_all('h3'):
    tmp=[]
    category=d.find(class_='tag').text
    title=d.find_all('a')[-1].text    
    tmp=news.get(category,[])
    tmp.append(title)
    news[category]=tmp
for category in news.keys():
    print(category, '新聞:')
    for title in news[category]:
        print(title)
    print('-----------------------------------------------------------------')
```

### ③ 哪一個類別的文章最多

```python
for category in sorted(news, key=lambda x:len(news[x]), reverse=True):
    print(category, len(news[category]))
        
```
